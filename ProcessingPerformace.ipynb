```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import nibabel as nib
%matplotlib inline
```

```python
dicom_df = pd.read_csv('/pylon5/ac5616p/debdas/SCCORProcessing/SCCOR_batch_df.csv', header=None)
dicom_df.head()
```

```python
success = 0
total_cases = len(dicom_df)
file_sizes = []
for i in range(total_cases):
    path = dicom_df[1][i]+'/'+dicom_df[1][i].split('/')[-1]+'.nii.gz'
    if os.path.isfile(path):
        success += 1
        file_sizes.append(os.stat(path).st_size)
        epi_img = nib.load(path)
        epi_img_data = epi_img.get_fdata()
        print(epi_img_data.shape)
```

```python
print(success)
```
