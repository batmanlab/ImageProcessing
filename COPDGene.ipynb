```python
import os
import pandas as pd
```

```python
phase1_path = '/pghbio/dbmi/batmanlab/chirayu/Data/COPDGene/Test_Reproducing_Phase-1_Images'
phase2_path = '/pghbio/dbmi/batmanlab/chirayu/Data/COPDGene/Images'
print("No. of patients in Phase-1",len(os.listdir(phase1_path)))
print("No. of patients in Phase-2",len(os.listdir(phase2_path)))
```

```python
i = 0
target_img_df = []
for phase1_patient in os.listdir(phase1_path):
    phase1_patient_path = os.path.join(phase1_path,phase1_patient,"Phase-1")
    phase1_img_types = os.listdir(phase1_patient_path)
    if ("Isotropic" in phase1_img_types) and ("SuperVoxels" in phase1_img_types):
        phase1_isotropic_path = os.path.join(phase1_patient_path, "Isotropic")
        phase1_supervoxel_path = os.path.join(phase1_patient_path, "SuperVoxels")
        phase1_isotropic = ""
        phase1_supervoxel = ""
        phase1_fhog = ""
        phase1_haralick = ""
        for phase1_isotropic_file in os.listdir(phase1_isotropic_path):
            matches = [phase1_patient+"_INSP_", "BSpline_Iso1.0mm.nii.gz"]
            if all(x in phase1_isotropic_file for x in matches):
                phase1_isotropic = os.path.join(phase1_isotropic_path,phase1_isotropic_file)
                break
        for phase1_supervoxel_file in os.listdir(phase1_supervoxel_path):
            matches = [phase1_patient+"_INSP_", "_BSpline_Iso1.0mm_SuperVoxel_Param30mm.nii.gz"]
            if all(x in phase1_supervoxel_file for x in matches):
                phase1_supervoxel = os.path.join(phase1_supervoxel_path,phase1_supervoxel_file)
                break
        for phase1_supervoxel_file in os.listdir(phase1_supervoxel_path):
            matches = [phase1_patient+"_INSP_", "_Iso1.0mm_SuperVoxel_Param30mm_fHOG_Hist_Features.h5"]
            if all(x in phase1_supervoxel_file for x in matches):
                phase1_fhog = os.path.join(phase1_supervoxel_path,phase1_supervoxel_file)
                break
        for phase1_supervoxel_file in os.listdir(phase1_supervoxel_path):
            matches = [phase1_patient+"_INSP_", "_BSpline_Iso1.0mm_SuperVoxel_Param30mm_haralick.h5"]
            if all(x in phase1_supervoxel_file for x in matches):
                phase1_haralick = os.path.join(phase1_supervoxel_path,phase1_supervoxel_file)
                break
                
        if (phase1_isotropic != "") and (phase1_supervoxel != "") and (phase1_fhog != "") and (phase1_haralick != ""):
            if phase1_patient in os.listdir(phase2_path):
                phase2_patient_path = os.path.join(phase2_path,phase1_patient,"Phase-2")
                phase2_img_types = os.listdir(phase2_patient_path)
                if ("Isotropic" in phase2_img_types) and ("SuperVoxels" in phase2_img_types):
                    phase2_isotropic_path = os.path.join(phase2_patient_path, "Isotropic")
                    phase2_supervoxel_path = os.path.join(phase2_patient_path, "SuperVoxels")
                    phase2_isotropic = ""
                    phase2_supervoxel = ""
                    phase2_fhog = ""
                    phase2_haralick = ""

                    for phase2_isotropic_file in os.listdir(phase2_isotropic_path):
                        matches = [phase1_patient+"_INSP_", "BSpline_Iso1.0mm.nii.gz"]
                        if all(x in phase2_isotropic_file for x in matches):
                            phase2_isotropic = os.path.join(phase2_isotropic_path,phase2_isotropic_file)
                            break
                    for phase2_supervoxel_file in os.listdir(phase2_supervoxel_path):
                        matches = [phase1_patient+"_INSP_", "_BSpline_Iso1.0mm_SuperVoxel_Param30mm.nii.gz"]
                        if all(x in phase2_supervoxel_file for x in matches):
                            phase2_supervoxel = os.path.join(phase2_supervoxel_path,phase2_supervoxel_file)
                            break
                    for phase2_supervoxel_file in os.listdir(phase2_supervoxel_path):
                        matches = [phase1_patient+"_INSP_", "_Iso1.0mm_SuperVoxel_Param30mm_fHOG_Hist_Features.h5"]
                        if all(x in phase2_supervoxel_file for x in matches):
                            phase2_fhog = os.path.join(phase2_supervoxel_path,phase2_supervoxel_file)
                            break
                    for phase2_supervoxel_file in os.listdir(phase2_supervoxel_path):
                        matches = [phase1_patient+"_INSP_", "_BSpline_Iso1.0mm_SuperVoxel_Param30mm_haralick.h5"]
                        if all(x in phase2_supervoxel_file for x in matches):
                            phase2_haralick = os.path.join(phase2_supervoxel_path,phase2_supervoxel_file)
                            break
                    if (phase2_isotropic != "") and (phase2_supervoxel != "") and (phase2_fhog != "") and (phase2_haralick != ""):
                        target_img_df.append((phase1_patient+"#Phase1",phase1_isotropic, phase1_supervoxel, phase1_fhog, phase1_haralick))
                        target_img_df.append((phase1_patient+"#Phase2",phase2_isotropic, phase2_supervoxel, phase2_fhog, phase2_haralick))
                        i += 1
                        if i > 20:
                            break
    
```
