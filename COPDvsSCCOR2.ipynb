```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img_df = pd.read_csv("fHOGSummaryCOPDSCCOR.csv")
img_df.head()
```

```python
img_df['File'][0]
```

```python
img_list = []
i = 0
for fHOG_file in img_df['File']:
    if img_df['Dataset'][i] == 'COPD':
        iso_img = fHOG_file.replace('/SuperVoxels/', '/Isotropic/')
        iso_img = iso_img.replace('_Iso1.0mm_SuperVoxel_Param30mm_fHOG_Hist_Features.csv.gz', '_BSpline_Iso1.0mm.nii.gz')
        iso_img = iso_img.replace('_STD_','_SHARP_')
        if os.path.exists(iso_img):
            img_list.append(iso_img)
        else:
            print(iso_img)
            
    elif img_df['Dataset'][i] == 'SCCOR':
        iso_img = fHOG_file.replace('_Iso1.0mm_SuperVoxel_Param30mm_fHOG_Hist_Features.csv.gz', '_Iso1.0mm.nii.gz')
        if os.path.exists(iso_img):
            img_list.append(iso_img)
    i += 1
```
